# EDA on Delay Limit

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Define the folder path
folder_path = './'

# Check if the folder exists and list the files
if os.path.exists(folder_path):
    files = os.listdir(folder_path)
    print(files)
else:
    print("Folder not found")
```

```python
# Function to perform descriptive statistics and visualization
def analyze_data(df, delay):
    if df.empty:
        print(f"No valid data for delay {delay}")
        return
    
    print(f"\nDescriptive Statistics for {delay}:")
    print(df.describe())

    # Histogram for Ping Time
    plt.figure(figsize=(10, 6))
    plt.hist(df['Ping Time (us)'], bins=20, edgecolor='k')
    plt.title(f'Histogram of Ping Time ({delay} Delay)')
    plt.xlabel('Ping Time (us)')
    plt.ylabel('Frequency')
    plt.show()


    # Histogram for Distance
    plt.figure(figsize=(10, 6))
    plt.hist(df['Distance (cm)'], bins=20, edgecolor='k')
    plt.title(f'Histogram of Distance ({delay} Delay)')
    plt.xlabel('Distance (cm)')
    plt.ylabel('Frequency')
    plt.show()

    # Combined Box Plot for Ping Time and Distance
    fig, ax = plt.subplots(1, 2, figsize=(18, 7))
    df.boxplot(column='Ping Time (us)', by='Delay', grid=False, ax=ax[0])
    ax[0].set_title('Ping Time (us)')
    ax[0].set_xlabel('Delay')
    ax[0].set_ylabel('Ping Time (us)')
    ax[0].tick_params(axis='x', rotation=45)
    
    df.boxplot(column='Distance (cm)', by='Delay', grid=False, ax=ax[1])
    ax[1].set_title('Distance (cm)')
    ax[1].set_xlabel('Delay')
    ax[1].set_ylabel('Distance (cm)')
    ax[1].tick_params(axis='x', rotation=45)

    fig.suptitle('Combined Box Plot of Ping Time and Distance by Delay')
    plt.show()

# Function to load and clean data
def load_and_clean_data(file_path):
    try:
        df = pd.read_csv(file_path)
        #df = df.dropna()
        #print(df.head())
        df['Ping Duration'] = pd.to_numeric(df['Ping Duration'], errors='coerce')
        df['Distance (cm)'] = pd.to_numeric(df['Distance (cm)'], errors='coerce')
        df['Ping Time (us)'] = pd.to_numeric(df['Ping Time (us)'], errors='coerce')
        df = df.dropna(subset=['Ping Duration', 'Distance (cm)', 'Ping Time (us)'])
        return df
    except Exception as e:
        print(f"Error loading {file_path}: {e}")
        return pd.DataFrame()
```

```python
def process_csv_files(folder_path):
    # Gather all CSV files with their delays
    files_with_delays = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            delay_str = filename.split('_')[3]
            if 'us' in delay_str:
                delay = int(delay_str.replace('us', ''))
            elif 'ms' in delay_str:
                delay = int(delay_str.replace('ms', '')) * 1000  # Convert milliseconds to microseconds
            files_with_delays.append((delay, file_path))
    
    # Sort files by delay in descending order
    files_with_delays.sort(reverse=True, key=lambda x: x[0])
    
    # Process each file
    for delay, file_path in files_with_delays:
        df = load_and_clean_data(file_path)
        delay_label = f"{delay}us" if delay < 1000000 else f"{delay//1000}ms"
        analyze_data(df, delay_label)


# Usage example
process_csv_files('./')
```

```python
def visualize_all_delays(folder_path):
    # Gather all CSV files with their delays
    files_with_delays = []
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            delay_str = filename.split('_')[3]
            if 'us' in delay_str:
                delay = int(delay_str.replace('us', ''))
            elif 'ms' in delay_str:
                delay = int(delay_str.replace('ms', '')) * 1000  # Convert milliseconds to microseconds
            files_with_delays.append((delay, file_path))
    
    # Sort files by delay in descending order
    files_with_delays.sort(reverse=True, key=lambda x: x[0])
    
    # Combine all data into one DataFrame
    combined_data = pd.DataFrame()
    for delay, file_path in files_with_delays:
        df = load_and_clean_data(file_path)
        df['Delay'] = f"{delay}us" if delay < 1000000 else f"{delay//1000}ms"
        combined_data = pd.concat([combined_data, df])

    # Ensure the Delay column is ordered from largest to smallest
    combined_data['Delay'] = pd.Categorical(combined_data['Delay'], categories=sorted(combined_data['Delay'].unique(), key=lambda x: int(x.replace('ms', '000').replace('us', '')), reverse=True))

    # Plotting the combined data
    if not combined_data.empty:
        plt.figure(figsize=(14, 7))
        combined_data.boxplot(column='Ping Time (us)', by='Delay', grid=False)
        plt.title('Box Plot of Ping Time by Delay')
        plt.suptitle('')
        plt.xlabel('Delay')
        plt.ylabel('Ping Time (us)')
        plt.xticks(rotation=45)
        plt.show()

        plt.figure(figsize=(14, 7))
        combined_data.boxplot(column='Distance (cm)', by='Delay', grid=False)
        plt.title('Box Plot of Distance by Delay')
        plt.suptitle('')
        plt.xlabel('Delay')
        plt.ylabel('Distance (cm)')
        plt.xticks(rotation=45)
        plt.show()

# Usage example
visualize_all_delays('./')
```
