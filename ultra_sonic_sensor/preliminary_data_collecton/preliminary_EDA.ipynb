# Preliminary Data Collection 


```python
import pandas as pd
pd.set_option('display.max_rows', None)

# Load the CSV file to examine the first few rows
file_path = 'Preliminary datacolletion ultrasonic sensor - ultrasonic data1.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Descriptive stats
df.describe()
```

Check nulls

```python
# Count the number of null values in each column of the dataset
null_counts = df.isnull().sum()
null_counts
```

```python
# Display rows that contain any null values to identify where the issues might be
null_data_rows = df[df.isnull().any(axis=1)]
null_data_rows.head()
```

```python
# Remove rows with any null values and check the new shape of the dataset
cleaned_df = df.dropna()
cleaned_data_shape = cleaned_df.shape
cleaned_data_shape
```

#### Group by analysis

```python
grouped_data = cleaned_df.groupby(["Sensor ID", "Range (cm)"])
descriptive_stats = grouped_data.describe()
```

```python
descriptive_stats = grouped_data.agg({
    'Distance (cm)': ['mean', 'std', 'min', 'max'],  # For the Distance column, calculate mean, std, min, and max
    'Echo Return Time (ms)': ['mean', 'std', 'min', 'max']      # For the Echo Return Time column, calculate mean, std, min, and max
})
```

```python
descriptive_stats
```

This analysis helps in understanding the consistency of sensor performance across different conditions or setups.

```python
# Step 1: Compute standard deviation for each sensor and range group
grouped_data = cleaned_df.groupby(["Sensor ID", "Range (cm)"])
distance_std = grouped_data['Distance (cm)'].std().reset_index(name='Distance Std')

# Step 2: Compute the standard deviation of the standard deviations for each sensor
std_of_std_per_sensor = distance_std.groupby('Sensor ID')['Distance Std'].std().reset_index(name='Std of Std')
std_of_std_per_sensor.sort_values("Std of Std",ascending=False)
```

Sensor std of std does tell us the varibility of variablity of distance for each sensors. AS we can see that the sensor id 4, 5, 6 does have a high varaibility. Sensor 9 doesn't record any of the input so it doesn't give any valuable insight for now.

## Visualize the data

### Analyze the Distance

This is to identify each sensor's range and distance that are not matching.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Get unique sensors and ranges
sensors = cleaned_df['Sensor ID'].unique()
all_ranges = sorted(cleaned_df['Range (cm)'].unique())

# Create a box plot for each sensor
for sensor in sensors:
    # Filter data for the current sensor
    sensor_data = cleaned_df[cleaned_df['Sensor ID'] == sensor]
    
    # Prepare the figure
    plt.figure(figsize=(14, 12))  # Larger figure size for better visibility and stretching
    
    # Create the boxplot
    sns.boxplot(x='Range (cm)', y='Distance (cm)', data=sensor_data, palette="Set2",
                showfliers=True, order=all_ranges)
    
    # Adding individual data points as scatter plot (optional)
    #sns.stripplot(x='Range (cm)', y='Distance (cm)', data=sensor_data, color='black', size=5, jitter=True)

    # Set titles and labels
    plt.title(f'Sensor ID {sensor} - Distance Measurements by Range for All Trials')
    plt.xlabel('Range (cm)')
    plt.ylabel('Distance (cm)')
    plt.xticks(range(len(all_ranges)), all_ranges)  # Set custom x-axis ticks based on all unique ranges
    
    # Set y-axis limits for more precision and length
    min_distance = sensor_data['Distance (cm)'].min() - 10
    max_distance = sensor_data['Distance (cm)'].max() + 10
    plt.ylim([min_distance, max_distance])
    
    # Set y-axis tick frequency for more detailed labels
    plt.gca().yaxis.set_major_locator(MultipleLocator(5))  # Adjust the value for desired tick spacing
    
    # Show plot
    plt.show()
```

These graph does help us to glence through these graph tells us general idea of how each sensor is measuring for each trial.

### Analysis on Echo Time


Echo time is directly related to the distance where it display similar trend for each sensors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetics for the plots
sns.set(style="whitegrid")

# Get unique sensors
sensors = cleaned_df['Sensor ID'].unique()

# Create a line plot for each sensor
for sensor in sensors:
    # Filter data for the current sensor
    sensor_data = cleaned_df[cleaned_df['Sensor ID'] == sensor]
    
    # Prepare the figure
    plt.figure(figsize=(10, 6))
    
    # Create the line plot
    sns.lineplot(x='Range (cm)', y='Echo Return Time (ms)', data=sensor_data, marker='o')
    
    # Add the reference line
    plt.plot([0, 40], [0, 40], color='red', linestyle='--', linewidth=2, label='Reference Line (y=x)')
    
    # Set titles and labels
    plt.title(f'Echo Return Time by Range for Sensor {sensor}')
    plt.xlabel('Range (cm)')
    plt.ylabel('Echo Return Time (ms)')
    
    # Add a legend for the reference line
    plt.legend()
    
    # Show plot
    plt.show()

```
